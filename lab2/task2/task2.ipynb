# Задание 1
**Выделить из полноцветного изображения каждый из каналов R, G, B  и вывести результат. Построить гистограмму по цветам (3 штуки).**

Клочкова

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
img1 = Image.open("animal.jpg")
img2 = Image.open("paint.jpg")
img3 = Image.open("fun.jpg")

# Конвертируем в RGB
img1 = img1.convert("RGB")
img2 = img2.convert("RGB")
img3 = img3.convert("RGB")

arr1 = np.array(img1)
arr2 = np.array(img2)
arr3 = np.array(img3)

fig, axes = plt.subplots(1, 3, figsize=(12, 6))
fig.suptitle('Исходные изображения', fontsize=16)

axes[0].imshow(arr1)
axes[0].set_title('Изображение 1')
axes[0].axis('off')

axes[1].imshow(arr2)
axes[1].set_title('Изображение 2')
axes[1].axis('off')

axes[2].imshow(arr3)
axes[2].set_title('Изображение 3')
axes[2].axis('off')

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(1, 4, figsize=(16, 4))
fig.suptitle('Выделение цветовых каналов - Image 1', fontsize=16)

axes[0].imshow(arr1)
axes[0].set_title('Исходное изображение')
axes[0].axis('off')

r_channel1 = arr1.copy()
r_channel1[:, :, 1] = 0  
r_channel1[:, :, 2] = 0  
axes[1].imshow(r_channel1)
axes[1].set_title('Красный канал (R)')
axes[1].axis('off')

g_channel1 = arr1.copy()
g_channel1[:, :, 0] = 0  
g_channel1[:, :, 2] = 0  
axes[2].imshow(g_channel1)
axes[2].set_title('Зеленый канал (G)')
axes[2].axis('off')

b_channel1 = arr1.copy()
b_channel1[:, :, 0] = 0 
b_channel1[:, :, 1] = 0 
axes[3].imshow(b_channel1)
axes[3].set_title('Синий канал (B)')
axes[3].axis('off')

# Гистограммы для каждого цветового канала Image 1
fig, axes = plt.subplots(2, 2, figsize=(15, 10))
axes = axes.ravel()

axes[0].hist(arr1[:, :, 0].ravel(), bins=256, range=(0, 255), color="red", alpha=0.7)
axes[0].set_title("Красный канал (R) - Image 1")
axes[0].set_xlabel("Интенсивность (0 - 255)")
axes[0].set_ylabel("Количество пикселей")

axes[1].hist(arr1[:, :, 1].ravel(), bins=256, range=(0, 255), color="green", alpha=0.7)
axes[1].set_title("Зеленый канал (G) - Image 1")
axes[1].set_xlabel("Интенсивность (0 - 255)")
axes[1].set_ylabel("Количество пикселей")

axes[2].hist(arr1[:, :, 2].ravel(), bins=256, range=(0, 255), color="blue", alpha=0.7)
axes[2].set_title("Синий канал (B) - Image 1")
axes[2].set_xlabel("Интенсивность (0 - 255)")
axes[2].set_ylabel("Количество пикселей")

axes[3].hist(arr1[:, :, 0].ravel(), bins=256, range=(0, 255), color="red", alpha=0.5, label="R")
axes[3].hist(arr1[:, :, 1].ravel(), bins=256, range=(0, 255), color="green", alpha=0.5, label="G")
axes[3].hist(arr1[:, :, 2].ravel(), bins=256, range=(0, 255), color="blue", alpha=0.5, label="B")
axes[3].legend()
axes[3].set_title("Все каналы - Image 1")
axes[3].set_xlabel("Интенсивность (0 - 255)")
axes[3].set_ylabel("Количество пикселей")

plt.tight_layout()
plt.show()
```
